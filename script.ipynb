## Scatter Matrices

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(grads[['Sample_size', 'Median']], figsize=(10,10), c='red')
```

```python
scatter_matrix(grads[['Sample_size', 'Median', 'ShareWomen']], figsize=(10,10), c='black')
```

We can see from the scatter plot matrix that ShareWomen-Median is moderately negatively linearly correlated. Therefore Majors that are more popular with women tend to pay less.

## Finding The Most Popular Majors

```python
grads[:10].plot.barh(x='Major',y='ShareWomen')
plt.title('Proportion Of Women In The Ten Highest Paying Majors')
plt.xlabel('Percentage')
plt.ylabel('Major')
```

```python
end = len(grads) - 10
grads[end:].plot.barh(x='Major',y='ShareWomen')
plt.title('Proportion Of Women In The Ten Lowest Paying Majors')
plt.xlabel('Percentage')
plt.ylabel('Major')
```

Women are under represented in the the highest paying majors, and over represented in the lowest paying majors.

```python
grads[:10].plot.barh(x='Major',y='Unemployment_rate')
plt.title('Proportion Of Unemployed In The Ten Highest Paying Majors')
plt.xlabel('Percentage')
plt.ylabel('Major')
```

```python
end = len(grads) - 10
grads[end:].plot.barh(x='Major',y='ShareWomen');
plt.title('Proportion Of Unemployed In The Ten Lowest Paying Majors')
plt.xlabel('Percentage')
plt.ylabel('Major')
```
